```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv(r'C:\Users\niyai\OneDrive\Desktop\kaggle\British_Airway_Review_cleaned.csv')
```

```python
df
```

```python
test = df.drop(columns = 'date')
```

```python
fig = px.line(
    df,
    title = "Number of Reviews",
    x = "date"
)
fig.update_traces(mode="markers+lines", hovertemplate=None)
fig.update_layout(hovermode="x")
```

```python
average_stars_by_month = round(df.groupby(['year','month'])['stars'].mean(),2).reset_index()
average_stars_by_month
```

```python
fig = px.line(
    average_stars_by_month,
    title = "Number of Reviews",
    x = "month",
    y = "stars",
    color = "year"
)
fig.update_traces(mode="markers+lines", hovertemplate=None)
fig.update_layout(hovermode="x")
```

```python
px.pie(
    df,
    title = "Number of Reviews",
    names = "seat_type"
)
```

```python
px.pie(
    df,
    title = "Number of Reviews",
    names = "type_of_traveller"
)
```

```python
fig = px.histogram(
    df,
    title = "Number of Reviews",
    x = "country"
)
fig.update_xaxes(categoryorder='total descending')
```

```python
starv = df.groupby(['stars', 'recommended']).size().reset_index(name='count')
starv
```

```python
fig = px.bar(
    starv,
    title = "Number of Reviews",
    x = "stars",
    y = "count",
    color = 'recommended',
    barmode = 'group',
    text_auto=True
)
fig.show()
```
